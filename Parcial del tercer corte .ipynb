```python
# Inciso A
from numpy import zeros, loadtxt,array
from cmath import exp, pi
from pylab import plot, show,xlim, title
Dnegocio= loadtxt("dow.txt",float)
plot(Dnegocio,'red')
show()
```

```python
#Inciso B
from numpy.fft import rfft
Coeficiente=rfft(Dnegocio)
print("\nLos coeficientes de la transformada de fourier discreta es : \n",Coeficiente)
plot(abs(Coeficiente), 'm')
xlim(-10,20)
title=('Transformada de fourier con los coeficientes')
show()
```

Aqui se halla los coeficiente de fourier utilizando la transformada de rapida , los cuales es una matriz compleja y se interpreta como lo datos periodicos de los negocios 

```python
# Inciso C
c=Coeficiente
m=len(Coeficiente)
g=int(m*0.1)
for i in range(0,m):
    if g >= i:
        c[i]=c[i]
    else:
        c[i]=0
```

Aqui se hace la reducción de la matriz de los coeficientes de fourier, lo cual utilizamos el 10% en este caso 
utilizamos el 51 de los elementos de esa matriz y lo demás es cero 

```python
#inciso d
from pylab import plot, show,xlim, xlabel,legend, title, ylabel 
from numpy.fft import irfft
M=irfft(c)
plot(Dnegocio,'c',label='Grafica principal')
plot(M,'red',label='Grafica inversa')
xlabel('Dias')
ylabel('valores ')
title(' Transformada inversa en 10%')
legend()
show()
```

Al aplicar la transformada inversa el ruido disminuye, dando una mejor observación del fenomeno, se tiene pero 
se hacer a un poco, podemos observar los dias que supuestamente más claro, pero aqui se tomo un parte de la matrices de los coeficientes de fourier.

```python
#inciso E
k=Coeficiente
r=len(Coeficiente)
g=int(r*0.02)
for i in range(0,r):
    if g >= i:
        k[i]=k[i]
    else:
        k[i]=0
t=irfft(k)
plot(Dnegocio,'c',label='Grafica principal')
plot(t,'k',label='Grafica inversa')
xlabel(' dias')
ylabel('valores ')
title(' Transformada inversa en 2%')
legend()
show()
```
